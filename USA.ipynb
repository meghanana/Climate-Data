```python
import pandas as pd
import matplotlib.pyplot as plt

#Read the CSV file
df = pd.read_csv("/Users/megh/code/project/Indicator_3_1_Climate_Indicators_Annual_Mean_Global_Surface_Temperature_577579683071085080.csv")

#This will only take the data from United States instead of every other country 
df_us = df[df['Country'] == 'United States']

#This shows the temmperature from 1961 to 2024
year_columns = [str(year) for year in range(1961, 2024)]
data = df_us[year_columns].T  

data = data.reset_index()
data.columns = ['Year', 'Value']
data['Year'] = data['Year'].astype(int)
data['Value'] = pd.to_numeric(data['Value'], errors='coerce')


plt.figure(figsize=(20, 6))  # Wider for better label spacing
plt.plot(data['Year'], data['Value'], linestyle='-', marker='o', color='red')


plt.title('United States: Climate Change (1961â2024)', fontsize=20)
plt.xlabel('Year')
plt.xticks(data['Year'], rotation=90)
plt.ylabel('Temerature (Â°F)')  
plt.grid(True, linestyle='.', alpha=1)
plt.tight_layout()
plt.show()
```
